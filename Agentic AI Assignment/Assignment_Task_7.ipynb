## Problem 1: Smart Light Controller (OOP)
**Industry:** IoT & Smart Home Automation  
**Objective:** Design a class-based system to model a smart light, allowing for state management (ON/OFF) and identity tracking for individual home devices.

```python
class SmartLight:
    def __init__(self, name):
        """Initializes the light with a name and a default status of OFF."""
        self.name = name
        self.status = "OFF"

    def turn_on(self):
        """Updates the status to ON."""
        self.status = "ON"

    def turn_off(self):
        """Updates the status to OFF."""
        self.status = "OFF"

    def display_status(self):
        """Prints the current state of the light."""
        print(f"{self.name} is {self.status}")

# Step 1: Create an instance of the SmartLight
my_light = SmartLight("Bedroom Light")

# Step 2: Perform the action based on sample input
my_light.turn_on()

# Step 3: Display the expected output
my_light.display_status()
```

## Problem 2: Employee ID Card System (OOP)
**Industry:** HR Technology & Identity Management  
**Objective:** Create a structured `Employee` class to store and display professional credentials, simulating the backend of a corporate HR management portal.

```python
class Employee:
    def __init__(self, name, emp_id, department):
        """
        Initializes the Employee object with personal and professional details.
        """
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_id_card(self):
        """
        Prints the formatted ID card details.
        """
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")

# Step 1: Input data from Sample
name_input = "Rahul"
id_input = "EMP102"
dept_input = "IT"

# Step 2: Create an instance of the Employee class
emp1 = Employee(name_input, id_input, dept_input)

# Step 3: Call the method to display the card
emp1.display_id_card()
```

## Problem 3: Mobile Contact Record (OOP)
**Industry:** Mobile App Development & Communication  
**Objective:** Implement a `Contact` class to model a digital phonebook entry, ensuring that personal names and communication data are correctly stored and retrievable.

```python
class Contact:
    def __init__(self, name, phone):
        """
        Initializes a new contact record with a name and phone number.
        """
        self.name = name
        self.phone = phone

    def display_contact(self):
        """
        Prints the contact details in the required format.
        """
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone}")

# Step 1: Input data from Sample
contact_name = "Anita"
phone_num = "9876543210"

# Step 2: Create an instance of the Contact class
new_contact = Contact(contact_name, phone_num)

# Step 3: Call the method to display the saved contact
new_contact.display_contact()
```

## Problem 4: Product Price Tag Generator (OOP)
**Industry:** Retail & E-commerce Logistics  
**Objective:** Develop a `Product` class to automate the generation of standardized price tags, ensuring consistent currency formatting and product identification.

```python
class Product:
    def __init__(self, name, price):
        """
        Initializes the product with a name and a numerical price.
        """
        self.name = name
        self.price = price

    def print_price_tag(self):
        """
        Prints the formatted price tag with the currency symbol.
        """
        print(f"Product: {self.name}")
        # Using the ₹ symbol to match the expected retail format
        print(f"Price: INR {self.price}")

# Step 1: Input data from Sample
prod_name = "Headphones"
prod_price = 2499

# Step 2: Create an instance of the Product class
item = Product(prod_name, prod_price)

# Step 3: Generate the price tag
item.print_price_tag()
```

## Problem 5: Movie Rating Display System (OOP)
**Industry:** Digital Media & Streaming Services  
**Objective:** Implement a `Movie` class to manage cinematic metadata, ensuring that user ratings are stored and displayed with a consistent numerical scale.

```python
class Movie:
    def __init__(self, name, rating):
        """
        Initializes the Movie object with a title and its user rating.
        """
        self.name = name
        self.rating = rating

    def display_details(self):
        """
        Prints the movie name and its rating on a 5-point scale.
        """
        print(f"Movie: {self.name}")
        # Standardizing the output to show the rating out of 5
        print(f"Rating: {self.rating} / 5")

# Step 1: Input data from Sample
movie_title = "Inception"
movie_rating = 4.8

# Step 2: Create an instance of the Movie class
my_movie = Movie(movie_title, movie_rating)

# Step 3: Call the method to display the movie details
my_movie.display_details()
```

## Problem 6: Delivery Address Manager (OOP)
**Industry:** Logistics & Supply Chain Management  
**Objective:** Create a `Delivery` class to manage customer routing data, simulating the core data structure used in automated courier dispatch systems.

```python
class Delivery:
    def __init__(self, customer_name, address):
        """
        Initializes the delivery record with the recipient's name and location.
        """
        self.customer_name = customer_name
        self.address = address

    def print_delivery_details(self):
        """
        Prints the formatted delivery manifest.
        """
        print("Delivery Details")
        print(f"Customer: {self.customer_name}")
        print(f"Address: {self.address}")

# Step 1: Input data from Sample
cust_name = "Suman"
delivery_loc = "Hyderabad"

# Step 2: Create an instance of the Delivery class
shipment = Delivery(cust_name, delivery_loc)

# Step 3: Call the method to display the details
shipment.print_delivery_details()
```
