## Problem 1: Smart Parking Lot Management System
**Industry:** Smart City Infrastructure & IoT  
**Objective:** Develop a monitoring function to track vehicle flow, calculate current occupancy, and trigger capacity alerts for urban parking management.

```python
def manage_parking(capacity, logs):
    """
    Calculates current occupancy from logs and validates against capacity.
    """
    currently_parked = 0

    # Step 1: Iterate through the logs to update the count
    for log in logs:
        if log.upper() == "IN":
            currently_parked += 1
        elif log.upper() == "OUT":
            currently_parked = max(0, currently_parked - 1)

    # Step 2: Determine status based on capacity
    if currently_parked >= capacity:
        status = "Full"
    else:
        status = "Available"

    return currently_parked, status

# Input as per sample
parking_capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

# Execution
count, current_status = manage_parking(parking_capacity, vehicle_logs)

# Step 3: Display Output in the required format
print(f"Currently Parked Vehicles: {count}")
print(f"Parking Status: {current_status}")
```

## Problem 2: Online Food Delivery Time Estimator
**Industry:** E-commerce & Logistics  
**Objective:** Build a predictive function that calculates the base delivery time and applies cumulative delays based on environmental factors like traffic density and weather severity.

```python
def estimate_delivery_time(distance, traffic, weather):
    """
    Calculates estimated delivery time (ETA) based on distance and external factors.
    Base Speed: 5 minutes per km.
    """
    # Step 1: Calculate Base Time
    # Industry standard assumption: ~5 mins per km for urban delivery
    base_time = distance * 5

    # Step 2: Apply Traffic Delays
    traffic_delay = 0
    if traffic.lower() == "high":
        traffic_delay = 10
    elif traffic.lower() == "medium":
        traffic_delay = 5

    # Step 3: Apply Weather Delays
    weather_delay = 0
    if weather.lower() == "rainy":
        weather_delay = 5
    elif weather.lower() == "stormy":
        weather_delay = 15

    # Step 4: Summing the total ETA
    total_eta = base_time + traffic_delay + weather_delay

    return total_eta

# Input from Sample
dist = 8
traffic_lvl = "High"
weather_cond = "Rainy"

# Execution
eta = estimate_delivery_time(dist, traffic_lvl, weather_cond)

# Step 5: Display Output in the required format
print(f"Estimated Delivery Time: {eta} minutes")
```

## Problem 3: Movie Theatre Seat Occupancy Analyzer
**Industry:** Entertainment & Event Management  
**Objective:** Calculate the utilization rate of a cinema hall and provide status alerts based on booking density to assist in scheduling decisions.

```python
def analyze_theatre_occupancy(total_seats, booked_seats_list):
    """
    Calculates occupancy percentage and determines show status.
    booked_seats_list is a list where each entry represents a booked seat.
    """
    # Step 1: Count total booked seats
    # len() gives the number of entries in the list
    num_booked = len(booked_seats_list)

    # Step 2: Calculate occupancy percentage
    occupancy_pct = (num_booked / total_seats) * 100

    # Step 3: Determine Show Status and Suggestions
    if occupancy_pct == 100:
        status = "Housefull"
    elif occupancy_pct >= 70:
        status = "Almost Full"
    else:
        status = "Seats Available"

    return occupancy_pct, status

# Input from Sample
total_capacity = 200
# Simulating a list with 150 entries as per problem description
booked_data = [1] * 150

# Execution
percent, show_status = analyze_theatre_occupancy(total_capacity, booked_data)

# Step 4: Display Output in the required format
print(f"Occupancy: {percent:.0f}%")
print(f"Show Status: {show_status}")
```

## Problem 4: Cloud Server Load Classification System
**Industry:** Cloud Infrastructure & DevOps  
**Objective:** Develop a monitoring function to aggregate server CPU readings, calculate the mean load, and classify the system's operational health into Normal, Warning, or Critical states.

```python
def classify_server_load(cpu_readings):
    """
    Calculates the average CPU usage and returns the server's health status.
    """
    if not cpu_readings:
        return 0, "No Data"

    # Step 1: Calculate average CPU usage using a loop
    total_load = 0
    for load in cpu_readings:
        total_load += load

    avg_load = total_load / len(cpu_readings)

    # Step 2: Classify status based on the defined rules
    if avg_load < 50:
        status = "Normal"
    elif 50 <= avg_load < 80:
        status = "Warning"
    else:
        status = "Critical"

    return int(avg_load), status

# Input from Sample
cpu_data = [45, 60, 70, 85, 90]

# Execution
average_cpu, server_status = classify_server_load(cpu_data)

# Step 3: Display Output in the required format
print(f"Average CPU Load: {average_cpu}%")
print(f"Server Status: {server_status}")
```

## Problem 5: Smart Classroom Resource Usage Monitor
**Industry:** Smart Education & Energy Management  
**Objective:** Create a monitoring function to scan device usage hours and flag resources that exceed the 8-hour daily limit for maintenance and energy alerts.

```python
def monitor_classroom_usage(usage_dict):
    """
    Scans a dictionary of classroom resources and flags those used for more than 8 hours.
    """
    overused_resources = []
    energy_alert = "No"

    # Step 1: Define the overuse threshold (8 hours)
    threshold = 8

    # Step 2: Iterate through the dictionary to check usage
    for resource, hours in usage_dict.items():
        if hours > threshold:
            overused_resources.append(resource)
            energy_alert = "Yes"

    # Step 3: Format the output
    # Joining multiple resources with a comma if more than one exists
    overused_str = ", ".join(overused_resources) if overused_resources else "None"

    return overused_str, energy_alert

# Input from Sample
resource_usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

# Execution
overused, alert = monitor_classroom_usage(resource_usage)

# Step 4: Display Output in the required format
print(f"Overused Resources: {overused}")
print(f"Energy Alert: {alert}")
```

## Problem 6: Online Event Registration Capacity Controller
**Industry:** Event Management & EdTech  
**Objective:** Develop a registration logic that caps confirmed entries at a fixed capacity and automatically redirects overflow participants to a waitlist.

```python
def manage_event_registration(capacity, total_registrations):
    """
    Allocates registration slots and determines the number of waitlisted users.
    """
    # Step 1: Logic to separate confirmed and waitlisted users
    if total_registrations > capacity:
        confirmed = capacity
        waitlisted = total_registrations - capacity
        status = "Closed"
    else:
        confirmed = total_registrations
        waitlisted = 0
        status = "Open"

    return confirmed, waitlisted, status

# Input from Sample
event_max = 100
total_applied = 105

# Execution
conf, wait, reg_status = manage_event_registration(event_max, total_applied)

# Step 2: Display Output in the required format
print(f"Confirmed Registrations: {conf}")
print(f"Waitlisted Users: {wait}")
print(f"Registration Status: {reg_status}")
```
